# 1. WAP to SWAP between two variables

```python
num1 = 2.5
num2 = 0.36
temp = num2
num2 = num1
num1 = temp

print(f"The value of num1 is: {num1}")
print(f"The value of num2 is: {num2}")
```

# How many seconds are there in 42 minutes 42 seconds?

```python
minutes = 42
seconds = 42
minutes_to_seconds = minutes * 60

total_seconds = minutes_to_seconds + seconds

print(f"There is {total_seconds} seconds in 42 minutes 42 seconds.")
```

# How many miles are there in 10 kilometers? (Hint: There are 1.609 kilometers in a mile.)

```python
kilometers = 10
total_miles = kilometers/1.609

print(f"There are {total_miles} miles in {kilometers} kilometers")
```
